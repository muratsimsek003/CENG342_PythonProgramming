#**Encapsulation**

Encapsulation is one of the fundamental concepts in object-oriented programming (OOP). It describes the idea of wrapping data and the methods that work on data within one unit. This puts restrictions on accessing variables and methods directly and can prevent the accidental modification of data. To prevent accidental change, an objects variable can only be changed by an objects method. Those types of variables are known as private variables.


A class is an example of encapsulation as it encapsulates all the data that is member functions, variables, etc.



image.png

A common feature in object oriented programming languages is that classes can usually hide their attributes from any prospective clients. Hidden attributes are usually called private. In Python this privacy is achieved by adding two underscores   **__**    to the beginning of the attribute name

image.png

**Example of Encapsulation** 

```python
class Employee:
  #constructor
  def __init__(self,name,salary,project):
    #data members
    self.name=name
    self.salary=salary
    self.project=project

  #method

  def show(self):
    #accessing public data member
    print("Name: ", self.name,"Salary: ",self.salary)
  
  #method
  def work(self):
    print(self.name, " is working on ", self.project)
```

```python
#creating object of a class
emp=Employee("Huseyin",8500,"Python Programming Project")
```

```python
emp.name
```

```python
emp.project
```

```python
emp.salary
```

```python
emp.show()
```

```python
emp.work()
```



*   Using encapsulation we can hide the internal representation of an object from the outside. This is called **information hiding**.
*   Also, encapsulation allows us to  **restrict direct access to variables and methods and prevent accidental data modification** by creating private data members and methods within a class.
*   Encapsulation is a way to restrict access to methods and variables from outside the class. When working with the class and dealing with sensitive data, it is not a good choice to have access to all variables used within the class
*   For example, let's say you have an attribute that is invisible from the outside of an object, and you combine it with methods that provide read or write access. In this case, you can hide certain information and control access to the internal state of the object. Encapsulation provides a way for us to access the required variable without giving the program full scope access to all variables of a class. This mechanism is used **to protect the data of one object from other objects**.

**Public method to access private members**

```python
class Employee:
  #contructor
  def __init__(self, name, salary):
    #public data member
    self.name=name
    #private member
    self.__salary=salary
  
  #public  instance method
  def show(self):
    #private members are accessible from a class
    print("Name: ", self.name, "Salary: ", self.__salary) 
```

```python
class Employee1:
  def __init__(self, name, salary):
    self.name=name
    self.__salary=salary

  def showhidden(self):
    print("private member:", self.__salary)
```

**Name Mangling to access private members**

We can directly access private and protected veriables from outside of a class 
through name mangling. The name mangling is created on an identifier by 
adding two leading underscores and one trailling underscore, like this **_classname__dataMember** where classname is the current class, and data member is the private variable name.

```python
class Employee:
  #contructor 
  def __init__(self, name, salary):
    #public data member
    self.name=name
    #private member
    self.__salary=salary
```

```python
emp=Employee("Muhsin", 10000)
```

```python
emp.name
```

```python
emp.salary
```

```python
emp.__salary
```

```python
emp._Employee__salary
```

```python
print("I can access the private member using Name Mangling: ", emp._Employee__salary)
```

**Getter and Setter in Python Encapsulation**

```python
class BankAccount:
  #constructor
  def __init__(self, name, money, address):
    self.name=name #public
    self.__money=money #private
    self.address=address #public

  #get and set method
  def getMoney(self):
    return self.__money

  def setMoney(self,amount):
    self.__money=amount

  #private method

  def __increase(self):
    self.__money=self.__money+500
```

```python
bank1=BankAccount("Mehmet Yua", 10000, "Ankara")
```

```python
bank1.name
```

```python
bank1.address
```

```python
bank1.__money
```

```python
bank1.getMoney()
```
