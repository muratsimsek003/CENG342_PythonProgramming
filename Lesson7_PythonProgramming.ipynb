#**Polymorphism**

**What is Polymorphism:** The word polymorphism means having many forms. In programming, polymorphism means the same function name (but different signatures) being used for different types.

Polymorphism is an ability (in OOP) to use a common interface for multiple forms (data types).

Suppose, we need to color a shape, there are multiple shape options (rectangle, square, circle). However we could use the same method to color any shape. This concept is called Polymorphism.


image.png

Polymorphism is a key concept in object-oriented programming that allows objects of different classes to be treated as if they were the same type of object. In Python, polymorphism is implemented through the use of method overriding and method overloading.

The advantages of polymorphism in Python's object-oriented programming include:

**Code Reusability: **Polymorphism allows you to reuse the same method or function with different inputs or objects. This can help reduce the amount of code you need to write, making your programs more efficient and easier to maintain.

**Flexibility:** Polymorphism allows you to write more flexible and adaptable code. By treating different objects as if they were the same type, you can create more generic functions and classes that can work with a wider variety of inputs and objects.

**Scalability:** Polymorphism allows you to easily add new types of objects or functions to your code as your program grows and evolves. This can help make your code more scalable and easier to maintain over time.

**Code Organization:** Polymorphism can help you organize your code more effectively by grouping related functions and classes together based on their shared behavior. This can make your code easier to understand and navigate, which can be especially useful in larger programs.

Overall, polymorphism is a powerful tool in Python's object-oriented programming that can help you write more efficient, flexible, and scalable code.

```python
class Employee:
  def salary_raisee(self):
    raise_rate=0.1
    result=100+100*raise_rate
    print("Employee:", result)

class CompEng(Employee):
  def salary_raisee(self):
    raise_rate=0.2
    result=100+100*raise_rate
    print("CompEng:", result)

class EEE(Employee):
  def salary_raisee(self):
    raise_rate=0.3
    result=100+100*raise_rate
    print("EEE:", result) 
```

```python
emp=Employee()
```

```python
ceng=CompEng()
```

```python
eee=EEE()
```

```python
employee_list=[emp,ceng,eee]
```

```python
for employee in employee_list:
  employee.salary_raisee()
```

##New Class

```python
from math import pi

class Shape:
  def __init__(self,name):
    self.name=name

  def area(self):
    pass

  def fact(self):
    return print("I am a two-dimmensional shape.")
  
  def __str__(self):
    return self.name

class Square(Shape):
  def __init__(self, length):
    super().__init__("Square")
    self.length= length

  def area(self):
    return print("square:", self.length**2)

  def fact(self):
    return print("Squares have each angle equal to 90 degrees")

class Circle(Shape):
  def __init__(self, radius):
    super().__init__("Circle")
    self.radius=radius

  def area(self):
    return print("circle:", pi*self.radius**2)
 
```

```python
square=Square(4)
circle=Circle(7)
```

```python
square.area()
```

```python
circle.area()
```

```python
list_shape=[square, circle]

for shape in list_shape:
 shape.area() 
```

```python
for shape in list_shape:
  shape.fact()
```

##New Class 

```python
class Vehicle:
  def desc(self):
    print("So many categories of vehicle")

  def wheels(self):
    print("Differs according to the vehicle category")


class Car(Vehicle):
  def wheels(self):
    print(4)

class Bus(Vehicle):
  def wheels(self):
    print(8)
```

```python
obj_vehicle=Vehicle()
obj_car= Car()
obj_bus=Bus()
```

```python
list_vehicle=[obj_vehicle,obj_car,obj_bus]
```

```python
for i in list_vehicle:
  i.wheels()
  i.desc()
```
